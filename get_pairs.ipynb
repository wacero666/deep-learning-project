```python
import numpy as np
import os
import sqlite3
import shutil
#https://labrosa.ee.columbia.edu/millionsong/sites/default/files/lastfm/demo_similars_db.py
```

```python
def get_list(data, tid, track_list, counter):
    data_unpacked = []
    src = "./images_matrix"
    src_files = os.listdir(src)
    pair_file_name = os.path.join(src, tid+".npy")
    
    for idx, d in enumerate(data.split(',')):
        
        if idx % 2 == 0:
            pair = [d]
        else:
            if pair[0] in track_list:
                full_file_name = os.path.join(src, pair[0]+".npy")
                print(full_file_name, "./pairs/%d/"%counter)
                if (os.path.isfile(full_file_name)):
                    path = "./pairs/%d/"%counter
                    os.makedirs(path)
                    shutil.copy(full_file_name, path)
                    shutil.copy(pair_file_name, path)
                    np.savetxt(path+'y.txt', np.array([float(d)]), fmt='%f')
                pair.append(float(d))
                data_unpacked.append(pair)
                counter += 1
    # sort
    data_unpacked = sorted(data_unpacked, key=lambda x: x[1], reverse=True)
    if len(data_unpacked) >10:
        for k in range(10):
            print (data_unpacked[k])
    print ('...')
    if len(data_unpacked) != 0:
        print ('(total number of pairs with such similarity value: %s' % len(data_unpacked))
    return data_unpacked, counter
```

```python
conn = sqlite3.connect('lastfm/lastfm_similars.db')
c = conn.cursor()
counter = 0
for root, dirnames, filenames in os.walk("./images_matrix"):
        
        for filename in filenames:
            filepath = os.path.join(root, filename)
            sql = "SELECT target FROM similars_src WHERE tid='%s'" % filename[:-4]
            res = conn.execute(sql)
            data = res.fetchone()
            if data != None:
                data_unpacked, counter = get_list(data[0], filename[:-4], [name[:-4] for name in filenames], counter)
#                 num_arr = np.load(filepath)
                
            
```
